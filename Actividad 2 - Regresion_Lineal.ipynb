```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x1= [1.0, 1.6, 3.4, 4.0, 5.2]
y1= [1.2, 2.0, 2.4, 3.5, 3.5]
```

```python
x1= np.array(x1, dtype=float)
y1=np.array(y1,dtype=float)

n=len(x1)
```

```python
xsum1=np.mean(x1)
ysum1=np.mean(y1)
xsum=np.sum(x1)
ysum=np.sum(y1)
x1y1=np.sum(x1*y1)
x2=np.sum(x1**2)
```

```python
a=(n*x1y1)-(xsum*ysum)
a1=(n*x2) - (xsum**2)
a2 = (a/a1)
```

```python
b= (ysum*x2)-(x1y1*xsum)
b1=(n*x2) - (xsum**2)
b2=(b/b1)
```

```python
y=a2+b2*x1
print(y)
```

```python
plt.scatter(x1,y)
plt.plot(x1,y)
```
